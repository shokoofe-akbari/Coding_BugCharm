```python
import sympy as sym
```

```python
x = sym.Symbol('x')
```

```python
sym.diff(x ** 5)
```

```python
sym.diff(x ** 2 * sym.cos(x ** (1/2)))
```

```python
sym.diff(1/(1+sym.exp(-x))) #sigmoid
```

```python
x, y = sym.symbols('x y')
```

```python
sym.diff(x ** 2 * y, x)
```

```python
x, y, z = sym.symbols('x y z')
```

```python
sym.diff(x ** 2 * y * z ** 2, z)
```

```python
sym.diff(x ** 2 * y * z ** 2, x)
```
