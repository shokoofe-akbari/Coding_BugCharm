```python
import numpy as np
import time  
import timeit
```

```python
my_list = list(range(1, 1_000_000))
my_np_list = np.array(my_list)
type(my_np_list)
```

```python
start_time = time.time()
s = 0
for i in range(len(my_list)):
    s += my_list[i]
end_time = time.time()
print(end_time - start_time)
```

```python
start_time = time.time()
sum_list = sum(my_list)
end_time = time.time()
print(end_time - start_time)
```

```python
start_time = time.time()    
sum_array = np.sum(my_np_list)
end_time = time.time()
print(end_time - start_time)
```

```python
%timeit sum_list = sum(my_list)
```

```python
%timeit sum_array = np.sum(my_np_list)
```

```python
def sum_list_func():
    sum(my_list)

def sum_array_func():
    np.sum(my_np_list)
```

```python
time_list = timeit.timeit(sum_list_func, number=100)
time_array = timeit.timeit(sum_array_func, number=100)

print(f"sum(my_list): {time_list:.6f}")
print(f"np.sum(my_np_list): {time_array:.6f}")
```
