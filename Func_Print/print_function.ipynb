### 2. Formatting Output: `sep` and `end` Arguments

```python
part1 = "Bug"
part2 = "Charm"
print(part1, part2, sep="       ")
```

```python
print("Hello World", end="\n")
print("Welcome to Python Programming.")
```

```python
day = 1
month = 1
year = 2026
print(day, month, year, sep="/")
```

```python
import time
print("Loading Simulation Data...", end="\t")
time.sleep(1) 
print("[Done]")
```

### 4. Direct File Output: Simple Logging

```python
log_message = "Simulation completed successfully at 14:00."

with open("bugcharm_log.txt", "w") as f:
    print(log_message, file=f)
    print("Error Rate: 0.01%", file=f)

print("Log saved to bugcharm_log.txt")
```

### 5. Modern Python: f-strings

```python
student_name = "Bug"
score = 98.5678
Id_num = 1500000

print(f"User: {student_name} | Score: {score}")
```

```python

print(f"Final Score (Rounded): {score:.10f}")
```

```python

print(f"Total ID: {Id_num:,}")
```

### 6. Printing in Color

```python
GRAY    = "\033[90m"
RED     = "\033[91m"
GREEN   = "\033[92m"
YELLOW  = "\033[93m"
BLUE    = "\033[94m"
PURPLE  = "\033[95m"
CAYAN   = "\033[96m"
BLACK   = "\033[30m"
RESET   = "\033[0m"  



print(f"Status: {GREEN}Online{RESET}")
print(f"Error:  {RED}Connection Lost{RESET}")

```

```python

cpu_temp = 85
print(f"Warning: CPU Temp is {YELLOW}{cpu_temp}°C{RESET} (High)")


BOLD    = "\033[1m"
UNDERLINE = "\033[4m"

print(f"\n{BOLD}Final Report:{RESET}")
print(f"Please click {UNDERLINE}{BLUE}Here{RESET} to download logs.")
```
