```python
site = 'BugCharm.com'
char = 'b'
model = 19.75
age = 29
print('this is my site: %s' %site)
print('this is a char: %c' %char)
print('this is a float: %.2f' %model)
print('this is a float: %.1f' %model)
print('this is a float: %f' %model)
print('this is a decimal: %d' %age)
print('this is a decimal: %i' %age)
```

```python
print('{}, is the best site.'.format(site))
```

```python
print('I am {} years old.'.format(age))
```

```python
a = '{} is the best site, There is {} videos on {}'
print(a.format(site, age, site))
```

```python
a = '{2} is the best site, There is {1} videos on {1}'
print(a.format(site, age, site))
```

```python
a = '{1} is the best site, There is {0} videos on {s}'
print(a.format(site, age, s='BugCharm'))
```

```python
a = 'site {0:20} {1:d}% is securty'
print(a.format('FaceBook', 100))
```

```python
print('My GPT is {:.2f}'.format(19.7854))
```

```python
print('{} was born in {} in {}'.format('BuCharm', 2024, 'University'))
```

```python
print('{:*^50s}'.format('BuCharm'))
print('{:=^50s}'.format('BuCharm'))
```

```python
name = "TMU"
print(f"|{name:<10}|")  # چپ‌چین
print(f"|{name:>10}|")  # راست‌چین
print(f"|{name:^10}|")  # وسط‌چین
print(f"| {'Name':<10} | {'Role':^10} |")
print(f"| {'Shokoofeh':<10} | {'Admin':^10} |")

text = "Log"
print(f"|{text:-^20}|") 
print(f"|{text:*^20}|") 

vesicles = 150
score = 98.5
print(f"vesicles = {vesicles}, score = {score}")
print(f"{vesicles=}, {score=}")

count = 10000000
print(f"Total: {count:,}") 
print(f"Total: {count:_}") 

change_a = 12.5
change_b = -3.2
print(f"Step A: {change_a:+}") # اجبار به نمایش مثبت
print(f"Step B: {change_b:+}")


prob = 0.8567
print(f"Success Rate: {prob:.1%}") 


num = 4500.75
integer = 42

print("--- Python Format Types ---")
print(f"Fixed Point (f):  {num:.2f}")  # استاندارد اعشاری
print(f"Scientific (e):   {num:.2e}")  # علمی
print(f"General (g):      {num:g}")    # هوشمند
print(f"Percentage (%):   {0.85:.1%}") # درصد
print(f"Symbol: {9733:c}") # کد 9733 ستاره است
print("-" * 30)
print(f"Binary (b):       {integer:b}")    # مبنای ۲
print(f"Hex (x):          {integer:x}")    # مبنای ۱۶
print(f"Zero Pad (03d):   {5:03d}")        # شماره‌گذاری

num = 1234.5678
print(f"|{num:-^15.2f}|") 

```
