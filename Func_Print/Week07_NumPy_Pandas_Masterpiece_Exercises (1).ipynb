## دیتاست‌های پایه

```python

import numpy as np
import pandas as pd

students = pd.DataFrame({
    "student_id": [101,102,103,104,105,106],
    "age": [22,24,23,29,26,21],
    "major": ["Physics","Math","Biology","Physics","Math","Biology"],
    "score_math": [18,15,np.nan,19,14,16],
    "score_stats": [17,16,14,np.nan,15,18]
})

lab = pd.DataFrame({
    "sample_id": [1,2,3,4,5,6],
    "temperature": [36.8,37.5,39.2,38.0,36.4,40.1],
    "pressure": [1.01,1.03,0.97,1.05,1.00,0.95],
    "device": ["A","A","B","B","A","B"]
})

sales = pd.DataFrame({
    "day": ["Sat","Sun","Mon","Tue","Wed","Thu"],
    "units": [120,180,90,210,160,130],
    "price": [10,10,12,11,10,11]
})

sensor_array = np.array([12.5,13.1,12.9,14.2,15.0,14.8,13.6])
```


## سؤال 1

### داستان
در سیستم آموزشی، نمرات ناقص باعث اختلال در تحلیل می‌شوند.

### خواسته‌ها

1. دانشجویانی که حداقل یک نمره ثبت‌نشده دارند را پیدا کنید.
2. میانگین نمرات هر دانشجو را محاسبه کرده و به دیتافریم اضافه کنید.



### راهنمایی
mean در Pandas به‌صورت پیش‌فرض NaN را نادیده می‌گیرد.

### پاسخ ناقص

```python
# کد خود را بنویسید
```

### پاسخ کامل

```python
students.assign(avg=students[['score_math','score_stats']].mean(axis=1))
```


## سؤال 2

### داستان
آزمایشگاه نیاز دارد دماهای بحرانی را سریع تشخیص دهد.

### خواسته‌ها

1. میانگین دما را محاسبه کنید.
2. نمونه‌هایی با دمای بالاتر از میانگین را استخراج کنید.



### راهنمایی
میانگین را یک بار حساب کنید و سپس فیلتر کنید.

### پاسخ ناقص

```python
# فیلتر دما
```

### پاسخ کامل

```python
lab[lab['temperature'] > lab['temperature'].mean()]
```


## سؤال 3

### داستان
مدیر فروش به دنبال روزهای با درآمد غیرعادی است.

### خواسته‌ها

1. درآمد روزانه را محاسبه کنید.
2. روزهایی با درآمد بالاتر از میانگین را نمایش دهید.



### راهنمایی
ستون جدید بسازید.

### پاسخ ناقص

```python
# total_income
```

### پاسخ کامل

```python
sales.assign(total=sales.units*sales.price).query('total > total.mean()')
```


## سؤال 4

### داستان
پایداری یک سنسور صنعتی بررسی می‌شود.

### خواسته‌ها

1. میانگین و انحراف معیار داده‌ها را محاسبه کنید.
2. مقادیر پرت را شناسایی کنید.



### راهنمایی
np.abs کمک می‌کند.

### پاسخ ناقص

```python
# outliers
```

### پاسخ کامل

```python
sensor_array[np.abs(sensor_array-sensor_array.mean())>sensor_array.std()]
```
