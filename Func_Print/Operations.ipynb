#### Membership Operation

```python
s = "BugCharm 01"
print('r' in s)
```

```python
seq = [1, 2, 3, 4, 5]
print(6 in seq)
print(3 in seq)
print('3' in seq)
```

```python
seq = [1, 2, 3, 4, 5]
print(6 not in seq)
print(3 not in seq)
```

#### Identity Operation

```python
x = 5
y = 5
print(x is y)
print(id(x) is id(y))
```

```python
x = [1, 3, 6]
y = [1, 3, 6]
print(x == y)
```

```python
x = [1, 3, 6]
y = [1, 3, 6]
print(x is y)
print(id(x))
print(id(y))
```

```python
x = [1, 3, 6]
y = [1, 3, 6]
print(x is not y)
print(id(x))
print(id(y))
```

```python
x = [1, 3, 6]
z = x
print(x is z)
print(id(x))
print(id(z))
```

```python
x = 6
z = x
print(x is z)
print(id(x))
print(id(z))
x = 10
print(x is z)
print(id(x))
print(id(z))
print(x)
print(z)
```
